```python
#importing pandas

import pandas as pd
```

```python
#loading dataset
df=pd.read_csv(r"C:\Users\dhima\Downloads\customer_data_null_values.csv")
df
```

```python
# 1. Select all columns for the customer with id 15.
df.iloc[5,:]
```

```python
# 2. Select the first 10 rows and the first 4 columns of the dataset.
df.iloc[1:11,1:5]
```

```python
# 3. Select 'customer_name' and 'purchase_amount' for customers in the North region.
df.loc[df["region"]=="North",["customer_name","purchase_amount"]]
```

```python
# 4. Select the last 15 rows and columns 3 through 6.
df.iloc[:,3:7].tail(15)
```

```python
df
```

```python

# 5. Select all columns for customers with purchase_amount greater than 200
df.loc[(df["purchase_amount"]>200),:]
```

```python
# 6. Select every third row from the first 30 rows and columns 0, 2, and 5.
df.iloc[0:30:3,[0,2,5]]
```

```python
# 7. Select 'customer_name', 'age', and 'occupation' for customers with gender 'Female'.
df.loc[(df["gender"]=="Female"),["customer_name","age","occupation","gender"]]
```

```python
# 8. Select the 20th to 25th rows (inclusive) and the last 4 columns.
df.iloc[[20,25],:].tail(4)
```

```python
# 9. Select all columns for customers enrolled in the loyalty program (loyalty_program = 'Yes').
df.loc[(df["loyalty_program"]=="Yes"),:]
```

```python
# 10. Select the first 5 columns for rows 10 to 20.
df.iloc[10:21,0:5]
```

```python
# 11. Select 'customer_name' and 'preferred_category' for customers with purchase_amount above 150.
df.loc[df["purchase_amount"]>150,["customer_name","preferred_category"]]
```

```python
# 12. Select rows 40 to 50 and columns 'customer_name', 'gender', and 'delivery_preference' by their
# index positions.
df.iloc[40:51,[1,3,9]]
```

```python
# 13. Select all columns for customers aged 25 who prefer 'Books' as their category.
df.loc[(df["preferred_category"]=="Books") & (df["age"]==25),:]
```

```python
# 14. Select the 6th column for the first 30 rows.
df.iloc[0:31,[6]]
```

```python
# 15. Select 'customer_name', 'purchase_amount', and 'region' for customers with delivery_preference 'Express'.
df.loc[df["delivery_preference"]=="Express",["customer_name","purchase_amount","delivery_preference"]]
```

```python
# 16. Select rows 5, 15, 25, 35, and 45 and columns 1, 4, and 7.
df.iloc[[5,15,25,35,45],[1,4,7]]
```

```python
# 17. Select all columns for customers whose names start with 'S'.
df.loc[df["customer_name"].str.startswith("S"),:]
```

```python
# 18. Select the last 8 rows and the first 3 columns.
df.iloc[-8,0:3]
```

```python
# 19. Select 'customer_name' and 'purchase_amount' for customers with purchase_amount between
# 100 and 200 (inclusive).
df.loc[df["purchase_amount"].between(100,200),["customer_name","purchase_amount"]]
```

```python
# 20. Select every 4th row from 0 to 60 and columns 3, 6, and 9.
df.iloc[0:60:4,[3,6,9]]
```

```python
# 21. Select 'customer_name', 'occupation', and 'loyalty_program' for customers aged between 20 and 30 (inclusive).
df.loc[(df["age"]>=20) & (df["age"]<=30),['customer_name', 'occupation','loyalty_program']]
```

```python
# 22. Select rows 60 to 70 and the columns corresponding to 'age', 'region', and 'preferred_category'.
df.loc[60:70,['age', 'region','preferred_category']]
```

```python
# 23. Select all columns for customers in the East region with purchase_amount less than 150.
df.loc[(df["region"]=="East") & (df["purchase_amount"]<150),:]
```

```python
# 24. Select the first 12 rows and the last 3 columns.
df.iloc[0:12,-3]
```

```python
# 25. Select 'customer_name', 'age', and 'delivery_preference' for customers with id in the range 30 to 50.
df.loc[(df["id"].between(30,50)),['customer_name', 'age', 'delivery_preference']]
```
